```python
with  open("names.txt", "r",  encoding="UTF-8") as file:
    print(type(file))
    print(file.readline(), end="")#取消print換行
    print(file.readline(), end="")
    print(file.readline(), end="")
# 離開縮排，自動close
file.closed
```

```python
with  open("names.txt", "r",  encoding="UTF-8") as file:
    data:str = file.read()
    print(type(data))
    print(data.split())

    names = data.split()
    # names:list[str] = data.split()
    print(type(names))
```

```python
with  open("names.txt", "r",  encoding="UTF-8") as file:
    data:str = file.read()
    names:list[str] = data.split()
print(len(names))

 # for 自訂變數 in 序列資料
for name in names:
    print(name)


```

```python
# 使用names[] ->subscript
print(names[0])
print(names[-1])
print(names[:5]) # slice，最前到5
print(names[-5:]) #-5到最後
```

```python
from csv import DictReader


with  open("aqx_p_488-2.csv", "r",  encoding="UTF-8", newline="") as file:
    reader:DictReader = DictReader(file)
    sites:list = list(reader)

for site in sites:
    print(site["sitename"], end=" ")
    print(site["county"], end=" ")
    print(site["aqi"], end=" ")
    print(site["status"])
```
